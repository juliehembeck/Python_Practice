```python
size_array = population.shape[0]
#Benford = np.zeros(size_array)
for x in range(size_array):
    y = str(x)
    population_list[x] = int(y[0])
```

```python
print(population_list)
```

```python
counts, bins = np.histogram(population_list)
pp.hist(bins[:-1], bins, weights=counts)
```

```python
pp.hist(population_list, density=True, bins=10)  # density=False would make counts
pp.ylabel('Probability')
pp.xlabel('Data');
```

```python
one = 0
two = 0
three = 0
four = 0
five = 0
six = 0
seven = 0
eight = 0 
nine = 0
for x in range(size_array):
    if population_list[x] == 1:
        one  +=  1
    elif population_list[x] == 2:
        two += 1
    elif population_list[x] == 3:
        three += 1
    elif population_list[x] == 4:
        four += 1
    elif population_list[x] == 5:
        five += 1
    elif population_list[x] == 6:
        six += 1
    elif population_list[x] == 7:
        seven += 1
    elif population_list[x] == 8:
        eight += 1
    elif population_list[x] == 9:
        nine += 1
```

```python
Benford_assortment = [one, two, three, four, five, six, seven, eight, nine]
Numbers = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(Benford_assortment)
```

```python
pp.bar(Numbers, Benford_assortment , color ='maroon',  
        width = 0.4) 
  
pp.xlabel("#") 
pp.ylabel("times # appeared") 
pp.title("Benford's Law") 
pp.show() 
```
