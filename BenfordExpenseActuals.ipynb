```python
import math
import collections

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
pd.options.display.max_rows = 16
```

```python
expenseactuals = pd.read_csv('Expense_Actuals.csv') #importing file with data
```

```python
expenseactuals
```

```python
all_funds = expenseactuals["ALL FUNDS"].tolist()
```

```python
all_funds
```

```python
city_fund = expenseactuals["CITY FUND"].tolist()
```

```python
city_fund
```

```python
size_array = len(city_fund)
for x in range(size_array):
    y = str(abs(city_fund[x]))
    city_fund[x] = int(y[0])
    
```

```python
one = 0
two = 0
three = 0
four = 0
five = 0
six = 0
seven = 0
eight = 0 
nine = 0
zero = 0
for x in range(size_array):
    if city_fund[x] == 1:
        one  +=  1
    elif city_fund[x] == 2:
        two += 1
    elif city_fund[x] == 3:
        three += 1
    elif city_fund[x] == 4:
        four += 1
    elif city_fund[x] == 5:
        five += 1
    elif city_fund[x] == 6:
        six += 1
    elif city_fund[x] == 7:
        seven += 1
    elif city_fund[x] == 8:
        eight += 1
    elif city_fund[x] == 9:
        nine += 1
    elif city_fund[x] == 0:
        zero += 1
```

```python
Benford_assortment = [one, two, three, four, five, six, seven, eight, nine]
Numbers = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(Benford_assortment)
print(sum(Benford_assortment))
```

```python
pp.bar(Numbers, Benford_assortment , color ='maroon',  
        width = 0.4) 
  
pp.xlabel("#") 
pp.ylabel("times # appeared") 
pp.title("Benford's Law") 
pp.show() 
```
